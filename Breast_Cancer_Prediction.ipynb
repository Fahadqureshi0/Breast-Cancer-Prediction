**Importing Dependecies**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data Collection and Preprocessing**

```python
# Laoding Datset into Pandas Dataframes:
cancer_data = pd.read_csv('/content/data (1).csv')
```

```python
print(cancer_data)
```

```python
# Getting 5 rows and columns:
cancer_data.head()
```

```python
# Encoding Diagnosis Column into Target:
label_encoder = LabelEncoder()
cancer_data['target'] = label_encoder.fit_transform(cancer_data['diagnosis'])
```

```python
cancer_data.tail()
```

```python
cancer_data.shape
```

```python
# Getting Information of Data:
cancer_data.info()
```

```python
# Checking for missing values:
cancer_data.isnull().sum()
```

```python
# Statistical Measures of Dataset:
cancer_data.describe()
```

```python
# Checking distribution of Target Values:
cancer_data['target'].value_counts()
```

```python
# 1 ==> Bengin
# 0 ==> Malignant
```

```python
cancer_data.head()
```

```python
cancer_data.groupby('target').mean()
```

**Seprating Features and Targets**

```python
X = cancer_data.drop(columns='target', axis=1)
Y = cancer_data['target']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting Training and Testing Data**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Trainnig Model**

**Logistic Regression**

```python
model = LogisticRegression(max_iter=10000)
```

```python
# Tarining the Logistic Regression Model:
model.fit(X_train, Y_train)
```

**Model Evaluation**

**Accuracy Score**

```python
# Accuracy Score of Trainnig Data:
training_data_accuracy = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, training_data_accuracy)
```

```python
print("Accuracy Score of Training Data: ", training_data_accuracy)
```

```python
# Accuracy Score of Test Data:
test_data_accuracy = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, test_data_accuracy)
```

```python
print("Accuracy Score of Testing Data: ", test_data_accuracy)
```

**Building Predictive System**

```python
input_data = (17.68,20.74,117.4,963.7,0.1115,0.1665,0.1855,0.1054,0.1971,0.06166,0.8113,1.4,5.54,93.91,0.009037,0.04954,0.05206,0.01841,0.01778,0.004968,20.47,25.11,132.9,1302,0.1418,0.3498,0.3583,0.1515,0.2463,0.07738)
# changing input data into array:
changing_input_data = np.asarray(input_data)
# reshaping numpy array as we predicting for one data point:
changing_input_data = changing_input_data.reshape(1,-1)

prediction = model.predict(changing_input_data)
print(prediction)

if (prediction[0]==0):
  print("The Breast Cancer is Malignant")
else:
  print("The Breast Cancer is Benign")
```
